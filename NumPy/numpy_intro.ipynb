```python
# normal distribution (density=True)
mean = 100
std = 10

sample = np.random.normal(loc=mean, scale=std, size=100)
plt.hist(sample, bins=30, edgecolor='black', linewidth=1.2, density=True) 

from scipy import stats
x=np.linspace(mean-4*std, mean+4*std, 1000)
plt.plot(x, stats.norm.pdf(x, loc=mean, scale=std))
```

#### Basic algebra

**Scalar product**

$u\cdot v = (u_{x}, u_{y}, u_{z})(v_{x}, v_{y}, v_{z}) = u_{x}v_{x} + u_{y}v_{y} + u_{z}v_{z} = |u| |v|\cos(\theta)$

```python
u = np.array([1,2,0])
v = np.array([2,3,1])
np.dot(u,v)
```

```python
A = np.random.randint(0,10,size=6).reshape(3,2)
B = np.random.randint(0,10,size=8).reshape(2,4)

print(f"A:\n{A}\n")
print(f"B:\n{B}")
```

```python
# Scalar product (3x2)(2,4)->(3x4)
np.dot(A,B)
```

```python
np.dot(B,A) # Order matters! (2,4)(3x2)->Error
```

**Cross product**

$u\times v = (u_y v_z - u_zv_y)\textbf{i} - (u_x v_z - u_zv_x)\textbf{j} + (u_x v_y - u_y v_x)\textbf{k} = |u| |v|\sin(\theta)$

```python
# Cross product (or vector product)
x = np.array([1,0,0])
y = np.array([0,1,0])
z = np.cross(x, y)
z
```

```python
# Plot vectors x, y and the cross product z

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
origin = [0,0,0]
ax.quiver(origin[0], origin[1], origin[2], x[0], x[1], x[2])
ax.quiver(origin[0], origin[1], origin[2], y[0], y[1], y[2])
ax.quiver(origin[0], origin[1], origin[2], z[0], z[1], z[2], color='red')
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
plt.show()
```

**Solve linear system**


$A\textbf{x} = b$


\begin{align}
\begin{bmatrix}
1&2&0\\
0&1&0\\
1&0&1\\
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2\\
z_3 
\end{bmatrix}
=
\begin{bmatrix}
1\\
0\\
2
\end{bmatrix}
\end{align}

```python
A = np.array([[1,2,0],[0,1,0],[1,0,1]])
b = np.array([1,0,2])
x = np.linalg.solve(A, b)
x
```
