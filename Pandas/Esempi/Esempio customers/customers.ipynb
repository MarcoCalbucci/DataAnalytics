#### Load dataframe

```python
df_orders = pandas.read_csv(r"ordered_items.tsv", sep="\t")
df_orders.head()
```

```python
# load dataframe
df_customers = pandas.read_csv(r"customers.tsv", sep="\t")
df_customers.head()
```

#### Merge

```python
# OLAP
df = df_orders.merge(df_customers, on=['customerid'])
df.head()
```

#### Save to csv

```python
# save to csv
df.to_csv(r"merged.csv", sep = ",", index = False)
```

#### Exercises

##### Esercizio 1 - Quali sono gli item acquistati da chi vive in Oregon?

```python
# Esercizio 1
# Quali sono gli item acquistati da chi vive in Oregon?
df[df['state'] == 'Oregon'][['item', 'quantity']]
#list(df[df['state'] == 'Oregon']['item'].values)
```

##### Esercizio 2 - Qual è l'item più popolare in Colorado?

```python
# Esercizio 2
# Qual è l'item più popolare in Colorado?
df[df['state'] == 'Colorado'][['item', 'quantity']].groupby('item').sum()
```

##### Esercizio 3 - In quale stato sono stati registrati più ordini?

```python
# Esercizio 3
# In quale stato sono stati registrati più ordini?
df.groupby(by=['state']).size()
```
