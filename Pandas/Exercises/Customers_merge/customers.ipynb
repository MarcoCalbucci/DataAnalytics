Merge the customers.tsv and ordered_items.tsv files and answer the questions below using pandas and excel pivot table.

```python
import pandas as pd
```

#### Load dataframe

```python
df_orders = pd.read_csv(r"https://raw.githubusercontent.com/MarcoCalbucci/DataAnalytics/main/Pandas/Exercises/Customers_merge/ordered_items.tsv", sep="\t")
df_orders.head()
```

```python
# load dataframe
df_customers = pd.read_csv(r"https://raw.githubusercontent.com/MarcoCalbucci/DataAnalytics/main/Pandas/Exercises/Customers_merge/customers.tsv", sep="\t")
df_customers.head()
```

```python
# df_orders
df_orders.info()
# df_orders.describe()
```

```python
# df_customers
df_customers.info()
# df_customers.describe()
```

#### Merge

```python
# OLAP
df = df_orders.merge(df_customers, on=['customerid'])
df
```

#### Save to csv


```python
# save to csv
df.to_csv(r"merged.csv", sep = ",", index = False)
```

#### Exercises

##### Exercise 1 - Which items were bought in Oregon?

```python
# Esercizio 1
# Quali sono gli item acquistati da chi vive in Oregon?
df[df['state'] == 'Oregon'][['item', 'quantity']]
#list(df[df['state'] == 'Oregon']['item'].values)
```

```python
df[df['state'] == 'Oregon'][['item', 'quantity']].sort_values(by='quantity', ascending=False)
```

##### Exercise 2 - Which is the most popular item in Colorado?

```python
# Esercizio 2
# Qual è l'item più popolare in Colorado?
df[df['state'] == 'Colorado'][['item', 'quantity']].groupby('item').sum()
```

##### Exercise 3 - In which state have been registered more orders?

```python
# Esercizio 3
# In quale stato sono stati registrati più ordini?
df.groupby(by=['state']).size()
```
