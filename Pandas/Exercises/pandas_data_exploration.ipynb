#### **1. Load the CSV data into a pandas data frame. Print some high-level statistical info about the data frame's columns. HINT: You may also want to rename the columns.**

```python
import pandas as pd
```

```python
df = pd.read_csv(concrete_url)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.rename(columns={
    'Cement (component 1)(kg in a m^3 mixture)': 'Cement',
    'Blast Furnace Slag (component 2)(kg in a m^3 mixture)': 'Blast Furnace Slag',
    'Fly Ash (component 3)(kg in a m^3 mixture)': 'Fly Ash',
    'Water (component 4)(kg in a m^3 mixture)': 'Water',
    'Superplasticizer (component 5)(kg in a m^3 mixture)': 'Superplasticizer',
    'Coarse Aggregate (component 6)(kg in a m^3 mixture)': 'Coarse Aggregate',
    'Fine Aggregate (component 7)(kg in a m^3 mixture)': 'Fine Aggregate',
    'Age (day)': 'Age',
    'Concrete compressive strength(MPa. megapascals)': 'Concrete compressive strength'
})
```

```python
df
```

#### **2. How many rows have a compressive strength > 40 MPa?**

```python
len(df[df['Concrete compressive strength(MPa. megapascals)']>40])
```
