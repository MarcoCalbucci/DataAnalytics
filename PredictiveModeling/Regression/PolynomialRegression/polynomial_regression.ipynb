```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = r'https://raw.githubusercontent.com/MarcoCalbucci/DataAnalytics/main/PredictiveModeling/Regression/PolynomialRegression/Data/Position_Salaries.csv'

df = pd.read_csv(url)
df
```

```python
X = df['Level'].to_numpy().reshape(-1,1)
y = df['Salary']
```

```python
# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Visualising the Linear Regression results
plt.scatter(X, y)
plt.plot(X, lin_reg.predict(X), color = 'red')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
## Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y)
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'red')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
