```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats


mean = 164
std = 7

x_min = mean - 4*std
x_max = mean + 4*std


x = np.arange(x_min, x_max, 0.1)
y = scipy.stats.norm.pdf(x, loc=mean, scale=std)
#y1 = scipy.stats.norm.cdf(x, loc=mean, scale=std)

plt.plot(x, y, label=f'mean = {mean}\nstd = {std}')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.fill_between(x[210:350], y[210:350], step="post", alpha=0.6)
plt.fill_between(x[140:420], y[140:420], color='blue',step="post", alpha=0.4)

plt.vlines(x=mean, ymin=0, ymax=0.06, linestyles='dashed')
```

```python
plt.plot(x, y, label=f'mean = {mean}\nstd = {std}')
plt.fill_between(x[400:], y[400:],step="post", alpha=0.6)
plt.fill_between(x[430:], y[430:], color='blue',step="post", alpha=0.4)
plt.vlines(x=mean, ymin=0, ymax=0.06, linestyles='dashed')
plt.xlabel('Height (cm)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

balance_list = [62000.00,
64000.00, 
49000.00, 
324000.00, 
1264000.00, 
54330.00, 
64000.00, 
51000.00, 
55000.00, 
48000.00, 
53000.00
]
df = pd.DataFrame({'Balance': balance_list})
df

plt.hist(df.Balance, bins=100, edgecolor='black')
plt.xlabel('Balance ($)')
plt.ylabel('Counts')
plt.vlines(x=df.Balance.mean(), ymin=0, ymax=7, linestyles='dashed', color='red', label= 'mean')
plt.vlines(x=df.Balance.median(), ymin=0, ymax=7, linestyles='dashed', color='green', label= 'median')
#plt.vlines(x=df.Balance.mode(), ymin=0, ymax=9, linestyles='dashed', color='purple', label= 'mode')
plt.legend()
plt.show()
```

```python
df.Balance.skew()
```
