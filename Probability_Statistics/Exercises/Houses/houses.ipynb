```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_product = pd.read_csv("https://raw.githubusercontent.com/MarcoCalbucci/DataAnalytics/main/Probability_Statistics/Exercises/Houses/Product.csv")
df_customer = pd.read_csv("https://raw.githubusercontent.com/MarcoCalbucci/DataAnalytics/main/Probability_Statistics/Exercises/Houses/Customer.csv")
```

#### Is the majority of customers male or female?

```python
gender_counts = df_customer['Gender'].value_counts()
gender_counts
```

```python
plt.bar(gender_counts.index, gender_counts)
```

#### House sold by customer state? (Pareto diagram)

```python
state_counts = df_customer['State'].value_counts()
state_counts
```

```python
fig, ax1 = plt.subplots()

plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax1.bar(state_counts.index, state_counts)
ax2.plot(state_counts.index, state_counts.cumsum()/state_counts.sum(), '-o', color='orange')
ax2.set_ylim(ymin=0)

plt.show()
```

#### What is the age of customers? (Histogram)

```python
# check for nan
df_customer.info()
```

```python
df_customer = df_customer.dropna(axis=0, how='any') # all
df_customer.info()
```

```python
df_customer
```

```python
# convert to numeric
df_customer['Age at time of purchase'] = pd.to_numeric(df_customer['Age at time of purchase'])
```

```python
print(f"mean: {df_customer['Age at time of purchase'].mean():.2f}")
print(f"median: {df_customer['Age at time of purchase'].median():.2f}")
print(f"mode: {df_customer['Age at time of purchase'].mode().values[0]}") # mode is a pandas Series!
```

```python
df_customer['Age at time of purchase'].hist(bins=20, edgecolor='black')
```

#### Is there a correlation between price and area? (Scatter plot)

```python
df_product.info() # first we explore df_product
```

```python
df_product
```

```python
# we need to convert the price to numeric

df_product['Price'] = df_product['Price'].str.replace('£', '')
df_product['Price'] = df_product['Price'].str.replace(',', '')

df_product['Price'] = pd.to_numeric(df_product['Price'])
```

```python
plt.scatter(df_product['Area (ft.)'], df_product['Price'])
plt.xlabel('Area (ft.)')
plt.ylabel('Price (£)')
```

```python
df_product[['Price', 'Area (ft.)']].corr()
```
