## Uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_uniform_hist(min, max, samples, bins, title='title', color="green"):
    "Simulate a Uniform Distribution"
    data = np.random.uniform(low=min, high=max, size=samples)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Frequency")
    values, bins, _ = plt.hist( data, bins=bins, alpha=0.5, edgecolor='black', color=color, label="Frequency")

    ax_bis = ax.twinx()
    values = np.append(values, 0)
    ax_bis.plot( bins, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, f'Mean: {data.mean():.2f}')

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data

interact(mc_uniform_hist,
        min=(0,100,1),
        max=(200,300,1),
        samples=(5000,6000,1),
        bins=(10,40,1),
         );
```

## Triangular distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_triangular_hist(min,mode,max,samples,bins,title='title',color="green"):
    "Simulate a Triangular Distribution"
    data = np.random.triangular(min, mode, max, size=samples)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Frequency")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, edgecolor='black', color = color,label = "Frequency")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1], color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, f'Mean: {data.mean():.2f}')

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data
    
interact(mc_triangular_hist,
        min=(0, 100,1),
        mode=(200,300,1),
        max=(400,500,1),
        samples=(5000,6000,1),
        bins=(10,40,1),
         );
```

## Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def mc_normal_hist(mean, std_dev, samples, bins,title='title',color="green"):
    "Simulate a Normal Distribution"
    data = np.random.normal(mean, std_dev, size = samples)
    fig = plt.figure(figsize=(15, 8))
    ax = plt.axes()
    plt.ylabel("Frequency")
    values, base, _ = plt.hist( data, bins = bins, alpha = 0.5, edgecolor='black', color = color, label = "Frequency")
    ax_bis = ax.twinx()
    values = np.append(values,0)
    ax_bis.plot( base, np.cumsum(values)/ np.cumsum(values)[-1],  color='darkorange', marker='o', linestyle='-', markersize = 1, label = "Cumulative Probability" )
    plt.xlabel('labels')
    plt.ylabel("Probability")

    plt.axvline(data.mean(), color='k', linestyle='dashed', linewidth=1)
    min_ylim, max_ylim = plt.ylim()
    plt.text(data.mean()*1.1, max_ylim*0.01, f'Mean: {data.mean():.2f}')

    plt.title(title)
    ax_bis.legend();
    ax.legend();
    plt.show()
    return data

interact(mc_normal_hist,
        mean=(100,200,1),
        std_dev=(10,50,1),
        samples=(10, 10000, 1),
        bins=(10, 100,1),
         );
```
